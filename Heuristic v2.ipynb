```python
import pandas as pd

# Load data from Excel file
file_path = 'nurse_schedule_project2_data_large_VA.xlsx'
locations = pd.read_excel(file_path, sheet_name='locations')
nurses = pd.read_excel(file_path, sheet_name='nurses')
patients = pd.read_excel(file_path, sheet_name='patients')
task_execution_time = pd.read_excel(file_path, sheet_name='task_execution_time')
distance_matrix = pd.read_excel(file_path, sheet_name='distance_matrix')
nurses['current_location'] = pd.NA
nurses['available_hours'] = 10

# Function to get travel time between two locations
def get_travel_time(from_location, to_location, distance_matrix):
    """
    Calculate the travel time between two locations based on the distance matrix.

    :param from_location: The starting location
    :param to_location: The destination location
    :param distance_matrix: DataFrame representing the distance matrix
    :return: Travel time in hours
    """
    # Assuming the distance matrix is structured with locations as both rows and columns
    distance = distance_matrix.loc[from_location, to_location]
    # Assuming a fixed travel speed to convert distance to time (e.g., 1 unit distance = 1 hour)
    travel_time = distance  # Modify this calculation based on actual speed and distance units
    return travel_time

# Modified function to assign nurses to tasks with travel time consideration
# Modified function to handle missing 'current_location'
def assign_nurses_with_travel(nurses, tasks, distance_matrix, max_hours_per_day=10):
    assignments = []
    for _, task in tasks.iterrows():
        for _, nurse in nurses.iterrows():
            # Check if 'current_location' column exists and nurse has a current location
            if 'current_location' in nurses.columns and pd.notna(nurse['current_location']):
                travel_time = get_travel_time(nurse['current_location'], task['location'], distance_matrix)
            else:
                travel_time = 0  # No travel time if 'current_location' is missing or not set

            total_time_required = task_execution_time['Time'] + travel_time
            
            if nurse['available_hours'] >= max(total_time_required):
                assignments.append({
                    'nurse': nurse['name'],
                    'task': task['name'],
                    'location': task['location'],
                    'hours': total_time_required
                })
                
                # Update nurse's available hours and current location if the column exists
                nurses.loc[nurses['name'] == nurse['name'], 'available_hours'] -= total_time_required
                if 'current_location' in nurses.columns:
                    nurses.loc[nurses['name'] == nurse['name'], 'current_location'] = task['location']
                break

    return pd.DataFrame(assignments)

assignments = assign_nurses_with_travel(nurses, patients, distance_matrix)

# Display the assignments
print(assignments)
```

```python
# Merge task execution time into the tasks DataFrame
# Assuming there is a common column (e.g., 'task_id') in both DataFrames to merge on
tasks_with_execution_time = pd.merge(tasks, task_execution_time, on='common_column', how='left')

# Updated function
def assign_nurses_with_travel(nurses, tasks, distance_matrix, max_hours_per_day=10):
    assignments = []
    for _, task in tasks.iterrows():
        for _, nurse in nurses.iterrows():
            if 'current_location' in nurses.columns and pd.notna(nurse['current_location']):
                travel_time = get_travel_time(nurse['current_location'], task['location'], distance_matrix)
            else:
                travel_time = 0  # No travel time if 'current_location' is missing or not set

            total_time_required = task['Time'] + travel_time  # 'Time' should be the name of the execution time column

            if nurse['available_hours'] >= total_time_required:
                assignments.append({
                    'nurse': nurse['name'],
                    'task': task['name'],
                    'location': task['location'],
                    'hours': total_time_required
                })
                nurses.loc[nurses['name'] == nurse['name'], 'available_hours'] -= total_time_required
                if 'current_location' in nurses.columns:
                    nurses.loc[nurses['name'] == nurse['name'], 'current_location'] = task['location']
                break

    return pd.DataFrame(assignments)

# Using the updated function with the merged tasks DataFrame
assignments = assign_nurses_with_travel(nurses, tasks_with_execution_time, distance_matrix)

# Display the assignments
print(assignments)
```

```python
from itertools import product
import numpy as np

excel_file_path = "nurse_schedule_project2_data_large_VA.xlsx"
excel_sheets = ['locations','nurses','patients','task_execution_time','distance_matrix']
sheets_full_data = {}
for sheet in excel_sheets:
    sheets_full_data[sheet] = pd.read_excel(excel_file_path, sheet_name=sheet)

sheets_full_data['locations'], sheets_full_data['nurses'], sheets_full_data['patients'], sheets_full_data['task_execution_time'], sheets_full_data['distance_matrix']
locations_df = sheets_full_data['locations']
nurses_df = sheets_full_data['nurses']
patients_df = sheets_full_data['patients']
task_time_df = sheets_full_data['task_execution_time']
distance_matrix_df = sheets_full_data['distance_matrix']

# Cleaning and preparing each dataframe

# 1. Locations - May not need much preparation as it's just a list of locations
locations = locations_df['Alexandria'].tolist()

# 2. Nurses - Splitting skillset into a list for easier processing
nurses_df['skillset'] = nurses_df['skillset'].str.split(', ')

# 3. Patients - Melting the dataframe to have one row per patient per day
patients_long_df = patients_df.melt(id_vars=['id', 'location'], 
                                    var_name='day', 
                                    value_name='task').dropna()

# 4. Task Execution Time - Ensure it's in a suitable format
task_time_dict = task_time_df.set_index('Task').to_dict()['Time']

# 5. Distance Matrix - Transforming it into a more usable format
distance_matrix_df.set_index('Unnamed: 0', inplace=True)
distance_matrix = distance_matrix_df.to_dict(orient='index')

# Showing the first few rows of the transformed dataframes for verification
nurses_df.head(), patients_long_df.head(), task_time_dict, list(distance_matrix.items())[0]

def calculate_travel_time(from_location, to_location, distance_matrix):
    """ Calculate travel time between two locations based on the distance matrix. """
    return distance_matrix[from_location][to_location]

def nurse_available(nurse, day, nurse_schedule):
    """ Check if a nurse is available on a given day. """
    return day not in nurse_schedule.get(nurse, [])

def assign_nurses_to_patients(nurses, patients, task_time, distance_matrix):
    """
    Assign nurses to patients based on skillset, location, and task requirements.
    This is a simplified version of the heuristic algorithm.
    """
    nurse_schedule = {}  # Dictionary to keep track of nurse assignments
    for location in locations:
        # Filter patients and nurses based on location
        local_patients = patients[patients['location'] == location]
        # Iterate over each day
        for day in ['M', 'T', 'W', 'Th', 'F', 'S', 'Su']:
            day_patients = local_patients[local_patients['day'] == day]
            for _, patient in day_patients.iterrows():
                task = patient['task']
                # Find a nurse with the required skillset who is available
                for _, nurse in nurses.iterrows():
                    if task in nurse['skillset'] and nurse_available(nurse['id'], day, nurse_schedule):
                        # Assign nurse to patient
                        nurse_schedule.setdefault(nurse['id'], []).append(day)
                        break  # Nurse found for this patient, move to the next patient

    return nurse_schedule

# Testing the function with the given data
nurse_schedule = assign_nurses_to_patients(nurses_df, patients_long_df, task_time_dict, distance_matrix)
nurse_schedule
```

```python
def assign_nurses_to_patients_v3(nurses, patients, task_time, distance_matrix):
    """
    Assign nurses to patients based on skillset, location, and task requirements,
    with the additional constraint that each nurse can work only for a maximum of 4 days a week.
    """
    nurse_schedule = {}  # Dictionary to keep track of nurse assignments
    nurse_patient_assignment = {}  # Dictionary to keep track of which patients are served by which nurse

    for location in locations:
        # Filter patients based on location
        local_patients = patients[patients['location'] == location]

        # Iterate over each day
        for day in ['M', 'T', 'W', 'Th', 'F', 'S', 'Su']:
            day_patients = local_patients[local_patients['day'] == day]

            for _, patient in day_patients.iterrows():
                task = patient['task']
                patient_id = patient['id']

                # Find a nurse with the required skillset who is available and has not exceeded 4 workdays
                for _, nurse in nurses.iterrows():
                    nurse_days_worked = len(nurse_schedule.get(nurse['id'], []))
                    if (task in nurse['skillset'] and 
                        nurse_available(nurse['id'], day, nurse_schedule) and 
                        nurse_days_worked < 4):
                        # Assign nurse to patient
                        nurse_schedule.setdefault(nurse['id'], []).append(day)
                        # Record the patient served by the nurse
                        nurse_patient_assignment.setdefault(nurse['id'], []).append(patient_id)
                        break  # Nurse found for this patient, move to the next patient

    return nurse_schedule, nurse_patient_assignment

# Testing the modified function with the given data
nurse_schedule_v3, nurse_patient_assignment_v3 = assign_nurses_to_patients_v3(nurses_df, patients_long_df, task_time_dict, distance_matrix)
nurse_schedule_v3, nurse_patient_assignment_v3
```

```python
def calculate_location_changes(nurse_patient_assignment, patients_df):
    """
    Calculate the number of times each nurse changes locations.
    """
    location_changes = {}
    for nurse, patient_ids in nurse_patient_assignment.items():
        # Track the locations visited by the nurse
        locations_visited = [patients_df[patients_df['id'] == pid]['location'].values[0] for pid in patient_ids]
        # Count the number of location changes
        location_changes[nurse] = sum(loc1 != loc2 for loc1, loc2 in zip(locations_visited, locations_visited[1:]))

    return location_changes

# Calculate location changes for each nurse based on the current assignments
location_changes = calculate_location_changes(nurse_patient_assignment_v3, patients_df)
total_location_changes = sum(location_changes.values())
total_location_changes
```
